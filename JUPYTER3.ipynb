```python
import pandas as pd
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 4, 5])

# read_data = pd.read_csv('C://Users/Ibirogbakehinde/Desktop/DATASCIENCE/products.csv', header =0, sep=',')
# x=read_data['unit_price']
# y=read_data['category_id']

# Perform linear regression using least squares
slope, intercept, r_value, p_value, std_err = linregress(x, y) 

# Predict y values using the linear equation
predicted_y = slope * x + intercept

# Plot the data and the linear regression line
plt.scatter(x, y, label='Actual data')
plt.plot(x, predicted_y, color='red', label='Linear Regression Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

# Print the slope and intercept
print(f'Slope: {slope}')
print(f'Intercept: {intercept}')
```
